```python
import warnings
warnings.filterwarnings('ignore')
import scipy.io
import numpy as np
import matplotlib.pyplot as plt

# Load the .mat file
data = scipy.io.loadmat('./allFaces.mat')

# Extract faces matrix and nfaces vector
faces = data['faces']
nfaces = data['nfaces'][0]
```

```python
n, m = 192, 168
n_rows = 32256
n_cols = faces.shape[1]

fig, axs = plt.subplots(6, 6, figsize=(10, 10))
fig.suptitle('First pictures of Person #1 to Person #36', fontsize=16)

person = np.zeros((n, m, 36))
for i in range(0, 36):
    curr_col = faces[:, np.sum(nfaces[:i])]
    person[:, :, i] = curr_col.reshape(m, n).T;
    person_img = person[:, :, i]
    axs[i // 6, i % 6].imshow(person_img, cmap='gray')
    axs[i // 6, i % 6].axis('off')

plt.show()
```
