```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image

# Load the images
image1 = np.array(Image.open('Recording-1.png'))
image2 = np.array(Image.open('Recording-2.png'))
```

```python
image2.shape
```

```python
plt.imshow(image1, cmap = 'gray')
plt.title('Recording 1')
plt.show()

plt.imshow(image2, cmap = 'gray')
plt.title('Recording 2')
plt.show()
```

```python
# stacking the two images into single array using vstack of np
X = np.vstack((image1.flatten(), image2.flatten()))
print(X.shape)
```

```python
# Here I am centering the image
mean = np.mean(X, axis=1, keepdims = True)
X = X - mean
```

```python
cov = np.dot(X,X.T)/(X.shape[1]-1)

U, S, V = np.linalg.svd(cov)
d = np.diag(1.0 / np.sqrt(S)) 

# whitening the image
whitener = np.dot(U, np.dot(d, U.T))
whitenedX = np.dot(whitener, X) 
```

```python
import numpy as np

def fastICA(signals,alpha=1, iterations=1000):
    m, n = signals.shape
    W = np.random.rand(m, m)

    for c in range(m):
        
        w = W[c, :].copy().reshape(m, 1)
        w /= np.sqrt((w ** 2).sum())

        for _ in range(iterations):
            w_old = w.copy()
            ws = np.dot(w.T, signals)

            # g(.) = 1/a log(cosh(ax)) is the constract function
            wg = (1 / alpha) * np.log(np.cosh(alpha * ws)).T

            # g'(.) = tanh(a.x)
            wg_ = np.tanh(alpha * ws)

            # updation of w
            w_new = (signals * wg.T).mean(axis=1) - wg_.mean() * w.squeeze()
            w_new -= np.dot(np.dot(w_new, W[:c].T), W[:c])
            w_new /= np.sqrt((w_new ** 2).sum())

            w = w_new

        W[c, :] = w.T
        
    return W
```

```python
W = fastICA(whitenedX, iterations=100)
```

```python
unMixed = whitenedX.T.dot(W.T)
```

```python
unmixed_iamge_1 = unMixed[:,0].reshape((image1.shape[0], image2.shape[1]))
```

```python
unmixed_iamge_2 = unMixed[:,1].reshape((image1.shape[0], image2.shape[1]))
```
